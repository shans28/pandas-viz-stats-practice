```python
# Import library
import pandas as pd
```

```python
# Read data
orders_data = pd.read_csv("https://raw.githubusercontent.com/shans28/pandas-viz-stats-practice/refs/heads/main/data/orders.csv")
```

```python
#Print sample data
orders_data.head()
```

```python
# Print data type of sample data read using pd
type(orders_data)
```

```python
#Print datatype of columns
orders_data.dtypes
```

```python
# Convert Date column to Date datatype
orders_data["Date"] = pd.to_datetime(orders_data["Date"])
```

```python
# Print  columns with data type of float
orders_data.select_dtypes(include='float64')
```

```python
# Calculate memory usage of pandas dataframe
orders_data.memory_usage()
```

```python
# Try soft conversion of dtypes
orders_data.infer_objects().dtypes
```

```python
#Print day from date column
orders_data["Date"].dt.day_name()
```

```python
# fill Type of Meal column with 20 chars length
orders_data['Type of Meal'].str.zfill(width=20)
```

```python
#Convert Type of Meal column to category
orders_data['Type of Meal']=orders_data['Type of Meal'].astype('category')
orders_data['Type of Meal'].dtype
```

```python
# Print how type of meal column as category is stored
orders_data['Type of Meal'].cat.codes
```

```python
# Count the unique value in Type of meal column
orders_data['Type of Meal'].value_counts()
```

```python
# estimate the price of meal by company Id
orders_data.groupby(['Type of Meal', 'Company Id' ]) ['Meal Price'].sum()
```
