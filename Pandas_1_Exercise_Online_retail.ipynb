```python
import pandas as pd
```

```python
#df = pd.read_csv("/content/sample_data/mnist_test.csv")
df = pd.read_excel("/data/Online_Retail.xlsx", engine="openpyxl")
```

```python
df.head(5)
```

```python
#Create a new column TotalPrice = Quantity × UnitPrice using .apply().
df['TotalPrice']=df.apply(lambda x: x['Quantity']*x['UnitPrice'], axis=1)
```

```python
df.shape
```

```python
#Filter all transactions where TotalPrice is above £1000.
df = df[df["TotalPrice"]>1000]
```

```python
df.shape
```

```python
#Sort the dataset by TotalPrice in descending order.
df=df.sort_values("TotalPrice", ascending=False)
```

```python
df
```

```python
#Group the data by Country and compute total Quantity and TotalPrice.
result  = df.groupby("Country").agg(Total_Quantity=("Quantity", "sum"),
    Total_Price=("TotalPrice", "sum")).reset_index()
```

```python
result
```

```python
#Top 5 customers by total spend.
top_customers = (df.groupby("CustomerID")["TotalPrice"].sum().sort_values(ascending=False).head(5))
top_customers
```

```python
#Extract the month from InvoiceDate and count transactions per month
df["Month"] =df["InvoiceDate"].dt.month
```

```python
#Filter transactions from the UK with positive Quantity.
transactions_per_month = df.groupby("Month")["InvoiceNo"].value_counts()
transactions_per_month
```

```python
#Create a column ValueLabel as 'High' if TotalPrice > 500 else 'Low'.
uk_filter=df[(df["Country"]=="United Kingdom") & (df["Quantity"]>0)]
```

```python
uk_filter.shape
```

```python
#Count number of High value transactions by country.
df["ValueLabel"]=df["TotalPrice"].apply(lambda x: "High" if x > 500 else "Low")
```

```python
df
```

```python
high_counts = (
    df[df["ValueLabel"] == "High"]
      .groupby("Country")["InvoiceNo"]
      .count()
      .reset_index(name="HighCount")    # clean index + rename column
)

print(high_counts)
```
