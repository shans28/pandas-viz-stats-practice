### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/Laxminarayen/Inceptez-batch-25-Classwork/refs/heads/main/Python-Class12-DescriptiveStatistics-Pandas/wind.data)

### Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.

```python
data = pd.read_csv(
    "https://raw.githubusercontent.com/shans28/pandas-viz-stats-practice/refs/heads/main/data/wind.data",
    sep="\s+",
    parse_dates=[[0, 1, 2]],
)

# Convert the 'Yr_Mo_Dy' column to datetime objects
data['Yr_Mo_Dy'] = pd.to_datetime(data['Yr_Mo_Dy'])

# Set the 'Yr_Mo_Dy' column as the index
data = data.set_index('Yr_Mo_Dy')

display(data.head())
display(data.dtypes)
```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
# Function to fix the year
def fix_year(date):
    if date.year > 2000:
        return date.replace(year=date.year - 100)
    return date

# Apply the function to fix the year
data.index = data.index.to_series().apply(fix_year)

display(data.head())
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
index_name = data.index.name
index_name
data.index.dtype
```

### Step 6. Compute how many values are missing for each location over the entire record.  
#### They should be ignored in all calculations below.

```python
display(data.shape)
data.isna().sum()
data=data.dropna()
```

### Step 7. Compute how many non-missing values there are in total.

```python
data.isna().sum()
data.shape[0]
```

### Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
#### A single number for the entire dataset.

```python
data.mean().mean()
```

### Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days

#### A different set of numbers for each location.

```python
# Calculate the min, max, mean, and standard deviations for each location
loc_stats = data.agg(['min', 'max', 'mean', 'std'])

# Transpose the DataFrame to have locations as rows and statistics as columns
loc_stats = loc_stats.T

display(loc_stats)
```

### Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

#### A different set of numbers for each day.

```python
day_stats=data.agg(['min', 'max', 'mean', 'std'])
day_stats = day_stats.T

display(day_stats)
```

### Step 11. Find the average windspeed in January for each location.  
#### Treat January 1961 and January 1962 both as January.

```python
# Find the average windspeed in January for each location
january_data = data[data.index.month == 1]
average_january_windspeed = january_data.mean()

display(average_january_windspeed)
```

### Step 12. Downsample the record to a yearly frequency for each location.

---



```python
#Downsample example
# Example time series (daily data)
rng = pd.date_range("2025-01-01", periods=10, freq="D")
sample_data = pd.Series(range(10), index=rng)

print("Original:")
print(sample_data)

monthly = sample_data.resample("M").mean()

print("\nDownsampled (Monthly Mean):")
print(monthly)
```

### Step 14. Downsample the record to a weekly frequency for each location.

```python
weekly_sample=data.resample("W").mean()
display(weekly_sample.head())
```

### Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
weekly = data.resample("W-MON", origin="1961-01-02")
stats = weekly.apply(lambda x: pd.Series({
    "min": x.min().min(),
    "max": x.max().max(),
    "mean": x.mean().mean(),
    "std": x.stack().std()
}))

stats = stats.head(52)
display(stats)
```
